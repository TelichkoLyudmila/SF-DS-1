## Бриф «Лаборатории исследований гражданского общества». Определение уязвимых групп населения

Согласно опросу «инФОМ» от декабря 2021 года, у 27 % россиян хватает
денег только на еду, а ещё 9 % не могут позволить себе полноценное
питание. Эти люди особенно внимательно следят за ценами, а темп роста
цен на продукты обычно превышает средний темп инфляции. При этом
Росстат считает, что расходы на продукты питания должны составлять
примерно 36 % от среднемесячных расходов россиянина (ещё около 10 %
приходится на услуги ЖКХ и жильё, 4 % — на лекарства).

До 2021 года «черта бедности» (жизнь на сумму ниже прожиточного
минимума) в России определялась стоимостью минимальной продуктовой
корзины. В том же году правительство «отвязало» уровень бедности от цен на
базовые продукты: с 2021 года прожиточный минимум рассчитывается как
44.2 % от медианного дохода граждан РФ за прошлый год.


### Задача:
- кластеризовать регионы России и определить, какие из них наиболее
остро нуждаются в помощи малообеспеченным/неблагополучным
слоям населения;
- описать группы населения, сталкивающиеся с бедностью;
- определить:
    - влияет ли число детей, пенсионеров и других социально уязвимых
групп на уровень бедности в регионе;
    - связаны ли уровень бедности/социального неблагополучия с
производством и потреблением в регионе;
    - какие ещё зависимости можно наблюдать относительно
социально незащищённых слоёв населения.

### 1.3. Сведение в единый датасет

Мы привели все регионы к единым названиям и далее применив маску на интересующие регионы, мы можем собрать единый датасет. Однако есть проблема, в том, что по сути мы имеем дело с признаками для кластеризации регионов и временными рядами одновременно. Можно было бы делать time series clusterisation по каждому признаку, в данном случае между регионами находятся закономерности: достигаются ли одинаковые значения во времени, есть ли схожие тенденции к одновременному увеличению и уменьшению или схожие повторяющиеся закономерности в разные годы. Т.к. у нас статистика погодовая, и временной ряд взят не очень длинный (с 2017 года - года смены методологии рассчета Росстата), то кластеризация временных рядом будет не очень продуктивна. Тогда будем рассматривать кластеризацию регионов по признакам.

Рассмотрим в данном случае кластеризацию по отдельным годам и кластеризацию на основе среднего признака за последние 3 года (2020-2022) эти годы охватывали пандемию и начало СВО, следовательно должны довольно явно выявить нуждающиеся регионы.

Сведем все датасеты в один, выделив год как признак. После в кластеризации будем использовать данный признак для фильтрации (в финальную кластеризацию он не войдет)

```python
# Напишем функцию, которая будет превращать широкий датасет в длинный и объединит все датасеты в один
def melting_datasets(dataframes_and_columns):
    result_df = None  # Пустой DataFrame для сохранения результатов
    i = 0
    for dataframe, column_name in dataframes_and_columns:  
        
        dataframe_tr = pd.melt(dataframe, id_vars=[column_name], var_name='Year', value_name=f'factor{i}').sort_values(by=[column_name, 'Year']).reset_index(drop=True) #, value_name=f'{dataframe}'
        #print(dataframe_tr)
        dataframe_tr['Year'] = dataframe_tr['Year'].astype(int)
        i += 1
        # Объединяем данные
        if result_df is None:
            result_df = dataframe_tr
        else:
            result_df = pd.merge(result_df, dataframe_tr, left_on=[result_df.columns[0], 'Year'], right_on=[dataframe_tr.columns[0], 'Year'], how='left') # on=[column_name, 'Year']
            
    # mask result_df
    return result_df
```

### 1.5. Датасет с относительными величинами

Отлично, за исключением столбцов со смертностью, у нас больше нет 0, т.е. скрытых пропусков. Прежде тем приступить к разбивке датасета по годам для дальнейшей кластеризации, сделаем еще одну вещь. В датасете есть признаки с номинальной величиной человек (например умерших детей или численность людей за линией бедности). В условиях нашей задачи кластеризации регионов мы можем как учитывать эффект масштаба региона, так и кластеризовать по относительным величинам. Т.е. очевидно, что в регионах с большей численностью населения будет и больше умерших и больше людей за чертой бедности. Если мы хотим иметь возможность выделить эти регионы как кластер (например чтобы доля соц помощи для них была больше из-за большей значимости региона в экономике страны) мы должны использовать данные в текущем виде. Но если нам надо понять, какова ситуация в регионе в независимости от размера, то эти данные должны быть относительными. Поэтому сделаем еще один датасет, где признаки в людях поделим на численность региона, чтобы получить относительный показатель. 

Поэтому в дальнейшем будем анализировать 2 датасета:
- с абсолютными или номинальными велчиинами
- с относительными величинами

```python
result_df_relative = result_df.copy()
# Число умерших на первом году жизни детей за год (человек) к общей численности населения региона в %
result_df_relative['child_mortality_rel'] = round((result_df_relative['child_mortality_rural'] + result_df_relative['child_mortality_urban'])/result_df_relative['avg_population']*100, 6)
# Общая численность инвалидов к численности населения региона в %
result_df_relative['disability_rel'] = round(result_df_relative['disability']/result_df_relative['avg_population']*100, 6)
# Численность населения с доходами ниже линии бедности к численности населения региона в %
result_df_relative['people_lower_poverty_line_rel'] = round(result_df_relative['people_lower_poverty_line']/result_df_relative['avg_population']*100, 6)
# Численность безработных граждан, зарегистрированных в государственных учреждениях службы занятости к численности населения региона в %
result_df_relative['unemployed_rel'] = round(result_df_relative['unemployed']/result_df_relative['avg_population']*100, 3)
# Всего зарегистрировано преступлений к численности населения региона в %
result_df_relative['crime_total_rel'] = round(result_df_relative['crime_total']/result_df_relative['avg_population']*100, 3)
# Выявлено лиц, совершивших преступления экономической направленности к численности населения региона в %
result_df_relative['crime_economical_rel'] = round(result_df_relative['crime_economical']/result_df_relative['avg_population']*100, 3)
# Выявлено лиц, совершивших преступления, не имеющих постоянного источника дохода к численности населения региона в %
result_df_relative['crime_without_income_rel'] = round(result_df_relative['crime_without_income']/result_df_relative['avg_population']*100, 3)
# Коэффициент естественного прироста = (Рождаемость - Смертность) / Общая численность населения в регионе
result_df_relative['natural_increase_rate'] = round(result_df_relative['natural_population_groth']/result_df_relative['avg_population']*100, 3)

result_df_relative = result_df_relative.drop(['child_mortality_rural', 'child_mortality_urban', 'disability', 'people_lower_poverty_line', 'unemployed', 'crime_total', 'crime_economical', 
                         'crime_without_income', 'natural_population_groth', 'avg_population'], axis =1)
result_df_relative.head()
```

### 2.4. Регионы, имеющие проблемы в большинстве признаков

До кластеризации хочется понять, какие регионы по каким признакам "лидируют". Т.к. по итогам нам будет наобходимо выбрать регионы, нуждающиеся в наибольшей поддержке, то хорошо бы сразу выделить регионы, которые попадают нижние 10% рейтинга по показателям (мы ранее привели все признаки к правильным знакам, т.е. сейчас для всех работает правило "чем больше, тем лучше"). Напишем функцию, которая отсортирует каждый признак по убыванию, и выделит регионы, попавшие в нижние 10%. 

```python
# Переведем регионы в индексы
result_avg_20_22_rel = result_avg_20_22_rel.set_index('Regions')
result_avg_20_22 = result_avg_20_22.set_index('Regions')
result_avg_20_22.head(2)
```

```python
def create_ranked_dataset(input_df):
    # Создаем словарь для хранения списков признаков для каждого региона
    region_features = {region: [] for region in input_df.index}

    # Перебираем признаки
    for feature in input_df.columns:
        # Сортируем регионы по текущему признаку
        sorted_regions = input_df.sort_values(by=feature, ascending=True)
        # Вычисляем количество регионов, соответствующих нижним 10%
        num_regions_10_percent = int(0.1 * len(input_df))
        # Получаем список регионов, попавших в нижние 10%
        selected_regions = sorted_regions.head(num_regions_10_percent).index
        # Добавляем признак к списку для каждого региона
        for region in selected_regions:
            region_features[region].append(feature)

    # Преобразуем словарь в DataFrame
    result_df = pd.DataFrame(list(region_features.items()), columns=['Region', 'Features'])

    # Сортируем регионы по количеству признаков
    result_df['NumFeatures'] = result_df['Features'].apply(len)
    result_df = result_df.sort_values(by='NumFeatures', ascending=False).reset_index(drop=True)

    return result_df



# Вызываем функцию для датасета с абсолютными значениями
problem_regions = create_ranked_dataset(result_avg_20_22.iloc[:, :18])

# Выводим результат
problem_regions.head(15)
```

У нас сюда попали Москва и СПб из-за высоких абсолютных показателей, поэтому посмотрим лучше относительные

```python
# Вызываем функцию для датасета с относительными значениями
problem_regions_rel = create_ranked_dataset(result_avg_20_22_rel.iloc[:, :17])

# Выводим результат
problem_regions_rel.head(15)
```

Отлично, мы выделили предположительно наиболее проблемные регионы. Далее в кластеризации посмотрим, в какие кластеры они попадут. Проведем еще одно интересное исследование: те регионы, которые попали в наш ТОП-15 (может варьироваться в зависимости от задач тендера) предположительно должны иметь больше всего проблем. По каким признакам они лидируют.

```python
top_15_regions = problem_regions_rel[:15]

# Создаем словарь для хранения количества упоминаний каждого признака
feature_counts = {}

# Перебираем признаки в топ-15 регионах
for features_list in top_15_regions['Features']:
    for feature in features_list:
        feature_counts[feature] = feature_counts.get(feature, 0) + 1

sorted_feature_counts = dict(sorted(feature_counts.items(), key=lambda item: item[1], reverse=True))
sorted_feature_counts
```

## 4. Кластеризация
### 4.1. K-means

Переходим к выбору метода кластеризации. Начнем с наиболее популярного

В качестве маркера качетсва кластеризации будем использовать коэффициент силуэта. Поскольку далее мы его будем тоже использовать, обернем сочетание K-means + коэф. силуэта в функцию

Датасет с относительными величинами (X_rel)

```python
def k_means_silhouette (X):
    silhouette_res = []
 
    for cluster_num in range(3, 11):
        k_means =  KMeans(n_clusters=cluster_num, init='k-means++', n_init=10, random_state=42)
        k_means.fit(X)
    # подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
        silhouette = silhouette_score(X, k_means.predict(X))
        silhouette_res.append((cluster_num, silhouette, len(set(k_means.labels_))) )
    df_k_means = pd.DataFrame(silhouette_res, columns=['cluster_num', 'silhouette', 'number_of_clusters'])
    result = df_k_means.sort_values(by = 'silhouette')
    return result
```

### 4.3. DBSCAN

Проверим кластеризацию еще одним "методом соседей" DBSCAN, он хорошо работает в условиях наличия выбросов, и может сформировать более четкие границы кластеров.

```python
def dbscan_silhouette(X):
    siliuette_list_dbscan = []
 
    for eps in np.arange(0.1,10,0.2):
        for min_sample in range(1,20,4):
            dbscan = DBSCAN(eps=eps, min_samples=min_sample)
            dbscan.fit(X)
            sil_score = silhouette_score(X, dbscan.labels_, metric='euclidean')
            siliuette_list_dbscan.append((eps, min_sample, sil_score, len(set(dbscan.labels_))) )
    df_dbscan = pd.DataFrame(siliuette_list_dbscan, columns=['eps', 'min_samples', 'sil_score', 'number_of_clusters'])
    result = df_dbscan.sort_values(by = 'sil_score')
    return result
```

```python
dbscan_silhouette(X_rel)
```

Метод DBSCAN падает с ошибкой в подсчете коэф.силуэта, что либо 85 кластеров, либо 1. Предположительно из-за достаточно разряженных данных на большом числе признаков он не может сформировать кластеры и либо определяет все значения в один кластер, либо в кластор "шумовые", обозначенный -1. Проверим это, воспользовавшись методом Nearest Neighbors. Он используется для вычисления расстояний до ближайших соседей, и значение eps выбирается, например, как 90-й процентиль этих расстояний.

```python
# Найдем оптимальное значение eps с использованием Nearest Neighbors
knn = NearestNeighbors(n_neighbors=2)
knn.fit(X_rel)
distances, _ = knn.kneighbors()
eps_value = np.percentile(distances[:, 1], 90)  # Пример: 90-й процентиль

# Задаем значения eps и min_samples
eps = eps_value # это значение 4.72
min_samples = 3

# Создаем и обучаем модель DBSCAN
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan.fit(X_rel)
```

```python
result_avg_20_22_rel['DBSCAN'] = dbscan.fit_predict(X_rel)
result_avg_20_22_rel['DBSCAN'].value_counts()
```

Да, как видно из результатов DBSCAN действительно определяет все данные в один кластер, а граничные точки определяет как шумовые. В таком виде нам этот метод не подходит.

```python
def Scatterplot_3D(dataset, model):
    
    # Создаем трехмерный scatter plot
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')

    # Разделяем данные по компонентам
    x = dataset[:, 0]
    y = dataset[:, 1]
    z = dataset[:, 2]

    # Цвета на основе меток кластеров
    colors = model.labels_

    # Создаем scatter plot с учетом меток кластеров
    scatter = ax.scatter(x, y, z, c=colors, cmap='viridis', marker='o', label='Observations')

    # Настройки графика
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.set_title('3D Scatter Plot with Clustering')

    # Показываем график
    plt.show()
```

```python
# 2018 год
dataset = result_2018_rel.set_index('Regions').drop(columns_to_exclude, axis=1)
X_rel = dataset.copy()
pipeline.fit_transform(X_rel)
clusters_all_years['2018'] = pipeline.named_steps['kmeans'].labels_
print("Распределение по кластерам: ", clusters_all_years['2018'].value_counts())

# 2019 год
dataset = result_2019_rel.set_index('Regions').drop(columns_to_exclude, axis=1)
X_rel = dataset.copy()
pipeline.fit_transform(X_rel)
clusters_all_years['2019'] = pipeline.named_steps['kmeans'].labels_
print("Распределение по кластерам: ", clusters_all_years['2019'].value_counts())

# 2020 год
dataset = result_2020_rel.set_index('Regions').drop(columns_to_exclude, axis=1)
X_rel = dataset.copy()
pipeline.fit_transform(X_rel)
clusters_all_years['2020'] = pipeline.named_steps['kmeans'].labels_
print("Распределение по кластерам: ", clusters_all_years['2020'].value_counts())

# 2021 год
dataset = result_2021_rel.set_index('Regions').drop(columns_to_exclude, axis=1)
X_rel = dataset.copy()
pipeline.fit_transform(X_rel)
clusters_all_years['2021'] = pipeline.named_steps['kmeans'].labels_
print("Распределение по кластерам: ", clusters_all_years['2021'].value_counts())

# 2022 год
dataset = result_2022_rel.set_index('Regions').drop(columns_to_exclude, axis=1)
X_rel = dataset.copy()
pipeline.fit_transform(X_rel)
clusters_all_years['2022'] = pipeline.named_steps['kmeans'].labels_
print("Распределение по кластерам: ", clusters_all_years['2022'].value_counts())
```

Видим, что в целом соблюдается правило, где один кластер с преобладающим числом регионов, один с минимальным и один средний, но по итогам кластеризации они были перемешаны, поэтому переназначим их по убыванию значений, попавших в кластер: 0, 1, 2

```python
# Функция для переназначения значений в столбце
def remap_clusters(column):
    value_counts = column.value_counts()
    return column.map({value_counts.idxmax(): 0, value_counts.idxmin(): 2}).fillna(1).astype(int)

# Применение функции ко всем столбцам
clusters_all_years_remapped = clusters_all_years.apply(remap_clusters)

# с 2017 и 2018 кластеры немного перепуталсиь, поправим
clusters_all_years_remapped['2017'] = clusters_all_years_remapped['2017'].replace({1: 2, 2: 1})
clusters_all_years_remapped['2018'] = clusters_all_years_remapped['2018'].replace({1: 2, 2: 1})
# Вывод результата
clusters_all_years_remapped  = clusters_all_years_remapped.reset_index()
clusters_all_years_remapped
```

Тут мы столкнулись с проблемой, что по сути имеем дело с признаками для кластеризации регионов и временными рядами одновременно. Однако т.к. у нас статистика погодовая, и временной ряд взят не очень длинный (с 2017 года - года смены методологии рассчета Росстата), то кластеризация временных была бы не очень продуктивна. Поэтому рассматривали кластеризацию регионов по признакам.

Для итоговой кластеризации были получены отдельные датасеты за 2017-2022 г, внутри каждого были таблицы Регионы\признаки. Также для основного анализирующего датасета (на котором проверялись методы кластеризации) был взят датасет на основе среднего признака за последние 3 года (2020-2022) эти годы охватывали пандемию и начало СВО, следовательно должны довольно явно выявить нуждающиеся регионы. Также учитывая, что некоторые признаки были в натуральном(абсолютном выражении) т.е. в людях, было очевидно, что будут сильно коррелировать признаки, завязанные на численность населения. Поэтому были отдельно созданы датасеты по каждому году с относительными признаками (т.е. деленые на численность населения).

Были выявлены следующие взаимозависимости между данными:
- между признаками "Доля нетрудоспособного населения" и "доля населения с доходами ниже линии бедности" кросс-корреляция выше 90%. Это говорит о том, что с большой вероятностью эти 2 признака связаны, ведь распределение регионов по ним практически идентично. Следовательно можно сделать вывод, что если в регионе большая доля стариков и детей это приводит к серьезной нагрузке для семей, что они оказываются за линией бедности в пересчете на человека. Безусловно такие регионы очень нуждаются в мерах социальной поддержки
- корреляцию 80% имеют признаки "отношение числа занятых в экономике региона к численности населения региона в трудоспособном возрасте" и "среднедушевые денежные доходы" - тоже выглядит логично т.к. чем больше людей имеющих возможность работать, работают, тем выше среднедушевой доход (поскольку в расчете этого показателя идет полная численность населения, а не численность работющего населения). Аналогичная высокая корреляция между рабочими и ВРП - принцип объяснения такой же (чем больше рабочих, тем выше ВРП на душу населения). Возможно, некоторые из этих признаков будут в дальнейшем исключены, т.к. они в целом объясняют регионы в схожем ключе
- менее 90%, но также высокая обратная корреляция между признаками "общая площадь жилых помещений, приходящаяся в среднем на одного жителя" и "коэффициент естественного прироста". Т.е. чем больше убыль населения, тем больше кв.м. приходится на 1 человека - выглядит закономерно. 
- интересная взаимосвязь с признаком welfare_expense_share (Удельный вес расходов на социальную политику в общем объеме расходов консолидированных бюджетов субъектов РФ): средне-высокая обратная корреляция с признаками среднедушевого денежного дохода, ВРП и доли работающего населения из трудоспособного (эти же показатели между собой тоже сильно скоррелированы). Можно предположить, что во многом социальная поддержка распределяется во многом основываясь на этой официальной статистике.
- Численность населения с доходами ниже линии бедности имеют достаточно высокую обратную корреляцию с оборотом розничной торговли на душу населения. Что в целом логично: чем меньше уровень заработка населения в регионе, тем меньший вклад в развитие региона (в частности торговлю) население вносит.
- Чем больше население, тем больше преступлений в натуральном выражении. Заметим, что средняя обратная корреляция наблюдается между этим признаком и всеми 3-мя видами престплений, что в целом тоже объяснимо учитывая минусовой естественный прирост.
